```python
# Let's import all needed libraries

import glob
import random
import base64
import pandas as pd

from PIL import Image
from io import BytesIO
from IPython.display import display, HTML

import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
# a function to store images in dataframe

pd.set_option('display.max_colwidth', -1)

def get_thumbnail(path):
    i = Image.open(path)
    i.thumbnail((150, 150), Image.LANCZOS)
    return i

def image_base64(im):
    if isinstance(im, str):
        im = get_thumbnail(im)
    with BytesIO() as buffer:
        im.save(buffer, 'jpeg')
        return base64.b64encode(buffer.getvalue()).decode()

def image_formatter(im):
    return f'<img src="data:image/jpeg;base64,{image_base64(im)}">'
```

```python
# read CSV file contains (eSports teams list) into a dataframe

es = pd.read_csv('./esports_teams.csv', sep=';')
#es
```

```python
# create a new column to store the image:

es['image'] = es.file_name.map(lambda f: get_thumbnail(f))
```
