```python
hera.groupby(['sponsor', 'game']).prize_total.sum()
```

```python
# Plot Teams in the 3rd place

plt.figure(figsize=[14.70, 8.27]);
sb.countplot(data = hera, x = 'sponsor', hue = 'game');
plt.xticks(rotation= 0);
plt.yticks([1, 2], [1, 2]);
plt.title('PLAYHERA Tournaments Sponsors per Game \n', fontdict=font);
plt.xlabel('\n Sponsor', fontdict=font);
plt.ylabel('COUNT \n', fontdict=font);
```

```python
plt.figure(figsize=[14.70, 8.27]);
hera.groupby(['sponsor']).prize_total.sum().plot();
plt.title('PLAYHERA Tournaments Sponsors Total Prizes \n', fontdict=font);
plt.xlabel('\n Sponsor', fontdict=font);
plt.ylabel('TOTAL PRIZES \n', fontdict=font);
plt.yticks([10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000], 
           ['10k', '15k', '20k', '25k', '30k', '35k', '40k', '45k', '50k']);
```
